#### Access Modifiers... Public , Private(2 underscores) & Protected(1 underscore)

```python
#class definition
class Car():

    def __init__(self, model):
        self.model = model          # public attribute
        self.battery = 90      # public attribute
###################################################################

# Main program
car1 = Car("Audi")
print(car1.model)
print(car1.battery)
        
```

```python
#class definition
class Car():

    def __init__(self, model):
        self.model = model          # public attribute
        self.__battery = 90      # private attribute

    def drive(self):
        print("I drive a",self.model)
        print("battery", self.__battery)
###################################################################

# Main program
car1 = Car("Audi")
print(car1.model)

        
```

```python
car1.drive()
```

```python
car1.__battery
```

### Inheritance (Parent - Child relationship)

#### Single Inheritance

```python
class Pclass():

    _a = 500

    def __init__(self):
        print('parent class constructor called')

    def _parmet(self):
        print('I am in Parent Method')

class Cclass(Pclass):

    b = 800

    def __init__(self):
        super(Cclass, self).__init__()     # invoking the parent class constructor
        print('child class constructor called also')

    def _chdmet(self):
        print('I am in Child Method')
        
```

```python
p = Pclass()
p._parmet()
```

```python
c = Cclass()
```

```python
c._parmet()
```

```python
c._a, c.b
```

```python
c._chdmet()
```

#### Multilevel Inheritance

```python
class GClass():
    __x = 100
    
    def _grandmet(self):
        print('I am in GF Method')
        
    # setter method
    def updatex(self, val):
        self.__x = val
    
    # getter method
    def printx(self):
        print('value of x is', self.__x)
        return self.__x
        

class Pclass(GClass):   
    a = 500
    
    def parmet(self):
        print('I am in Parent Method')
        
class Cclass(Pclass):  
    b = 800
    
    def chdmet(self):
        print('I am in Child Method')
```

```python
chdobj = Cclass()
chdobj._grandmet()
```

```python
chdobj.updatex(150)
```

```python
val = chdobj.printx()
print(val)
```

```python
chdobj.b = 1000
```

```python
chdobj.b
```

#### Multiple Inheritance

```python
class Father():
    a = 200
    def fatmet(self):
        print("Father method called !")
        
class Mother():
    b = 300
    def motmet(self):
        print("Mother method called !")
        
class Child(Father, Mother):
    c = 800
    def chdmet(self):
        print("Child method called !")
```

```python
chdobj = Child()
chdobj.fatmet()
```

```python
chdobj.motmet()
```

```python
chdobj.chdmet()
```

#### Hierarchical I nheritance

```python
class Person():

    def __init__(self, n, sn):
        self.name = n
        self.surname = sn

class Employee(Person):

    def __init__(self, n, sn, sal, ot):
        super(Employee, self).__init__(n, sn)
        self.salary = sal
        self.overtime = ot

class Customer(Person):
    phoneNo = 988888888
    emailAdd = 'abc@gmail.com'
```

```python
emp1 = Employee('anish', 'Mohanan', 100000, 10)
emp1.name
```

```python
cust1 = Customer('athira','Regi')
cust1.name, cust1.phoneNo
```

```python
# Parent class
class Vehicle:
    def __init__(self, make, model):
        self.make = make
        self.model = model

    def start_engine(self):
        return f"The engine of {self.make} {self.model} is starting."

# Child class 1
class Car(Vehicle):
    def start_engine(self):
        return f"The car {self.make} {self.model} engine is starting with a roar."

# Child class 2
class Motorcycle(Vehicle):
    def start_engine(self):
        return f"The motorcycle {self.make} {self.model} engine is starting with a vroom."

# Child class 3
class Truck(Vehicle):
    def start_engine(self):
        return f"The truck {self.make} {self.model} engine is starting with a rumble."

# Creating objects of each child class
car = Car("Toyota", "Camry")
motorcycle = Motorcycle("Harley-Davidson", "Street 750")
truck = Truck("Ford", "F-150")

# Calling the start_engine method on each object
print(car.start_engine())        # Output: The car Toyota Camry engine is starting with a roar.
print(motorcycle.start_engine()) # Out
print(truck.start_engine())
```

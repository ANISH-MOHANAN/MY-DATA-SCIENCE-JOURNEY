4.	Explore the relationship between the feature and target column using a count plot of categorical columns and a scatter plot of numerical columns

```python
### categorical variables
insurance.region.value_counts()
```

```python
##count plot
import warnings
warnings.filterwarnings('ignore')

sns.countplot(x='region',data=insurance,palette='plasma')
plt.show()
```

```python
insurance.sex.value_counts()
```

```python
##count plot
sns.countplot(x='sex',data=insurance,palette='plasma')
plt.show()
```

```python
insurance.smoker.value_counts()
```

```python
##count plot
sns.countplot(x='smoker',data=insurance,palette='plasma')
plt.show()
```

```python
### comparison between target ('charges') and a categorical variable ('region') 

sns.barplot(x='region',y='charges',data=insurance,estimator=np.std,palette='plasma')
plt.show()
```

```python
### we can infer from the above plot that southeast region contributes to the highest insurance charges.
```

```python
### comparison between target ('charges') and a categorical variable ('sex') 

sns.barplot(x='sex',y='charges',data=insurance,estimator=np.std,palette='plasma')
plt.show()
```

```python
### we can infer from the above plot that males contributes to the highest insurance charges.
```

```python
### comparison between target ('charges') and a categorical variable ('region') 

sns.barplot(x='smoker',y='charges',data=insurance,estimator=np.std,palette='plasma')
plt.show()
```

```python
### we can infer from the above plot that smokers contributes to the highest insurance charges.
```

#### # If - Else

```python
hippo_wt = 800
rihno_wt = 800

if (hippo_wt > rihno_wt ):
    print("Hippo weighs more than rihno")
    print("Hippo wins")
elif(hippo_wt < rihno_wt):
    print("Rihno weighs more than hippo")
    print("Rihno wins")
else:
    print("Hippo and rhino weighs same")
    print("It's a tie")
```

```python
name = input("Enter your name: ")
percent = int(input("Enter the percentage obtained: "))

if(percent >=60):
    print(name,' Passed with percentage: ', percent, ' and secured grade in first division')
elif(percent >= 50 and percent < 60):
    print(name, 'Passed with percentage: ', percent, ' and secured grade in Second Divison')
elif(percent >= 40 and percent < 50):
    print(name, 'Passed with percentage: ', percent, ' and secured grade in Third divison')
else:
       print('sorry to inform you that ', name,' has Failed')
```

#### While Loop... iterative statement

```python
i = 1                  # initialize the counter
while (i < 5):         # check for condition
    print(i)
    i = i + 1          # increment the counter
```

```python
i = 1
while (i <= 5):
    family_members = input("Enter the name of family members: ")
    print(family_members)
    i = i + 1
```

```python
# Create a list of brands of bikes . 
# the user should enter 5 name of 2-wheeler brands
# it should be stored in a list and printed once the input is completed by the user
# Hint :
# bike = [] ==> blank list
# bike.append(x)
```

```python
bike = []
for i in range (1 , 6, 1):
    brand = input("Enter the name of bike brands: ")
    bike.append(brand)

print("The bike brand is: ", bike)
```

```python
bikes = []

num = int(input("Enter the number of bikes that need to be stored :"))
for i in range(1, num + 1, 1):
    brand = input("Enter brand name : ")
    bikes.append(brand)
    
print('The brands entered are : ', bikes)
```

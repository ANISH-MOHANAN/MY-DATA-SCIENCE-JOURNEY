#### Recursive functions:

##### the functions that call themselves within the definition. Helps to break up a big job into smaller similar tasks

```python
def factorial(number):
    if number == 1:
        return 1
    else:
        return number*factorial(number-1)
        
```

```python
factorial(6)
```

#### Generator functions:

##### this function yeild values one at a time and generates a sequence of values using the yeild keyword.

```python
def my_gen():
    for x in range(2,10):
        yield x
```

```python
for i in my_gen():
    print(i)
```

```python
###the output of the my_gen function is stored in an object.---to get the values the object stores we have to use a for loop
### to print each value stored OR we use the next statement
```

```python
X=my_gen()
print(next(X))
print(next(X))
print(next(X))
print(next(X))
print(next(X))
print(next(X))
print(next(X))
print(next(X))
```

#### Differences-:
-  Yeild statement in a generator saves the state of a function. With the 'Next' statement the function is called again and the excecution happens from the next value(not the initial value,execution happens with the recent value).The return statement terminates the function completely.

-  The generators do not run the function it creates a state of the function and gives us a generator object,to get the value the object stores we have to use the next statement for a 'for in' loop

```python
#Q) create a generator function which will give us the odd numbers from the natural number list.
```

```python
def odd_numbers():
    n = 1
    yield n
    n = n + 2
    yield n
    n = n + 2
    yield n
```

```python
odd = odd_numbers()
print(next(odd))
print(next(odd))
print(next(odd))
```

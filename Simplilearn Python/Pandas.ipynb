#### Functions of a dataframe:

* tail,head-selectively displays the bottom and top 5 rows of a dataframe.
* sample() functions helps select random rows and columns from a dataframe.
* info() function gives more info on the dataset
* describe() functions gives the summary stats for the variables in the data
* query() functions helps us to query the dataset on some condition
* loc() helps us to query the dataset based on column and row labels
* iloc() helps us to query the dataset based on the index position of rows and columns
* unique() helps us get the distinct values in a certain row or column
* null() helps us to locate missing values
* fillna() helps us to fill missing values in the dataset with some values
* rename() helps us to rename the column name


**Histograms-- We use histograms to study the distribution of a continuous variable. By assumption all continuous variables follow a normal distribution.**

```python
x=np.random.normal(170,10,250)# we are generating 250 datapoints from a normal distribution with a mean of 170 and
                            # a SD of 10

plt.hist(x, color='red')
plt.show()
```

```python
y=np.array([35,25,25,15])
mylabels=['Daisy','Lilies','Marigold','Rose']
colors=['orange','purple','magenta','yellow']
plt.pie(y,labels=mylabels,autopct='%1.1f%%',colors=colors)
plt.show()
```


**Barplots- it helps us to study the distribution of a categorical variable. We have the categories along the x axis,and the frequency count along the y axis. Countplots from the seaborn package helps us do the same.**

```python
x=np.array(['Single','Married','Separated'])
y=np.array([25,32,15])

plt.bar(x,y,color='green')
plt.show()
```

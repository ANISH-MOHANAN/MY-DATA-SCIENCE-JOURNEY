```python
# Problem statement 1:
'''
Depending on the percentage obtained by a student give a grade :
Perc >= 60 ==> First Division
Perc between 50 & 59 ==> Second Division
Perc between 40 & 49 ==> Third Division
Perc less than 40 ==> Fail 
'''
```

```python
name = input('Enter Name:')
percent = int(input('Enter Percentage:'))

if(percent >= 60):
    print(name,' Passed with percentage: ', percent, ' and secured grade in first division')
elif(percent >= 50 and percent < 60):
    print(name,' Passed with percentage: ', percent, ' and secured grade in second division')
elif(percent >= 40 and percent < 50):
    print(name,' Passed with percentage:', percent,' and secured grade in third division')
else:
    print('sorry to inform you that ', name,' has Failed')
```

```python
# Problem statement 2:
'''
A company insures a driver when the following conditions are met :
1. If the driver is married
2. If the driver is un-married , male & above 30 yrs of age
3. Other than the above cases the driver will not get insurance
'''
```

```python
ms = input('Enter the marital status of driver (M / UM) :').upper()
sex = input('Enter the sex of the driver (M / F) : ').upper()
age = int(input('Enter age of the driver : '))

if (ms == 'M'):
    print("Driver can get an insurance")
elif (ms == 'UM' and sex == 'M' and age > 30):
    print("Driver can get an insurance")
else:
    print("We cannot give insurance")
```

```python
# Create a list of brands of bikes . 
# the user should enter 5 name of 2-wheeler brands
# it should be stored in a list and printed once the input is completed by the user
# Hint :
# bike = [] ==> blank list
# bike.append(x)
```

```python
bikes = []

num = int(input("Enter the number of bikes that need to be stored :"))
for i in range(1, num + 1, 1):
    brand = input("Enter brand name : ")
    bikes.append(brand)
    
print('The brands entered are : ', bikes)
```

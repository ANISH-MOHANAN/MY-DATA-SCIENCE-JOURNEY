- Multithreading is a concept in which we allow the CPU to execute multiple threads of a process at the same time.Job is work that needs to be done.A task is a piece of work that needs to be done. The process is a series of actions that is done for a particular purpose.Job and task define the work to be done, whereas process defines the way the work can be done or how the work should be done. It is like when several actions are done in a parallel manner on the same object. Two mechanics are working on the same car. If one changes the front tires, the other the rear ones. Now in programming concept we use the concept of multithreading when we say want to execute multiple functions(where each function performs a specific task) at the same time.If we execute the functions one by one, it will consume more time.


- Threading is a technique which allows the CPU to execute multiple threads of the same program at the same time to make the execution of the program faster.

- Thread is a predifned class which is available in the module. Threads are basic units of the CPU and it is known for independent execution. 


- Multithreading is used to run multiple tasks and functions of a certain program at the same time.

```python
#we craete a class 'First' and define the run() method inside the class. This method is the starting point
# of a thread and the implementation of the thread is defined within this method
# say i want to print a string 10 times and that is what i want the run() method to do.
class First:
    def run(self):
        for i in range(10):
            print("Multithreading, Thread1")
```

```python
#Now we define another class and use the same implementation Logic under 
#the run() method. We will run #both the methods at the same time using 
#multiple thread, so to identify the execution of each, we will #print a 
#different string
class Second:
    def run(self):
        for i in range(10):
            print("Multithreading, Thread2")
```

```python
# so now we have both the functions within the same program defined in two different classes
```

```python
# we will define two threads to execute each of these functions independently. 
```

```python
# we define threads 't1' and 't2' which are two threads belonging to the class 'First' and 'Second' respectively.
t1=First()
t2=Second()
```

```python
#calling the run method of both the objects/threads.
t1.run()
t2.run()
```

**Till now we have only one thread called 'Main Thread' which executes the entire program. This main thread first executes instance 't1' and then executes instance 't2'. When the main thread executes 't1' then the second instance 't2' waits. Now to understand the total time of execution, we import 'sleep' class from the 'time' module.**

We use 1 sec of sleep time for each thread,i.e while executing 't1' there is 1 second of wait time, and then 't2' is executed. Again while execution of 't2' we have a sleep time of 1 sec.

**So we get the same output,but each execution of either method happens with a time lag of 1 sec.**


```python
from time import sleep

class First:
    def run(self):
        for i in range(10):
            print('Multithreading, Thread1')
            sleep(1)
            
class Second:
    def run(self):
        for i in range(10):
            print('Multithreading, Thread2')
            sleep(1)

#we define threads 't1' and 't2' which are two threads belonging to class 'First' and 'Second' respectively. 
t1=First()
t2=Second()

#calling the run methods of both the objects/threads.
t1.run()
t2.run()
```

**If there is a time lag of 1 sec,the entire time taken by t1 is 10 secs, while the execution of t2 is 10 secs. Hence the total time required for the execution of the entire program by the main thread is 10+10+10=30 seconds**


• Now we will try to reduce the time taken, by including the concept of multithreading whereby each of the process is executed by two separate threads and not by a single 'Main Thread'.

**Now though we have imported the Thread class and both the classes have inherited the properties of the thread class,multithreading does not work as the program still takes 20 seconds to execute the whole program. To ensure it takes less time, we will call the start method. This method will call the run method from either thread and multithreading will work. The total time for execution will be 10 seconds.**





```python
#for doing this, we will import 'thread' class from the 'Threading' module. And let class First and second inherit all
#the properties of the thread class.

from time import sleep
from threading import Thread


class First(Thread):
    def run(self):
        for i in range(10):
            print('Multithreading, Thread1')
            sleep(1)

class Second (Thread): 
    def run(self):
        for i in range(10):
            print('Multithreading, Thread2')
            sleep(1)

#we define threads 't1' and 't2' which are two threads belonging to class 'First' and 'Second' respectively. 

t1=First()
t2=Second()

#Using start method to call the run method

t1.start() 
t2.start()


```

**The above program gets executed in 10 seconds and the output changes in every execution.**

```python
#for doing this, we will import 'thread' class from the 'Threading' module. And let class First and second inherit all
#the properties of the thread class.

from time import sleep
from threading import Thread


class First(Thread):
    def run(self):
        for i in range(10):
            print('Multithreading, Thread1')
            sleep(1)

class Second (Thread): 
    def run(self):
        for i in range(10):
            print('Multithreading, Thread2')
            sleep(1)

#we define threads 't1' and 't2' which are two threads belonging to class 'First' and 'Second' respectively. 

t1=First()
t2=Second()

#Using start method to call the run method

t1.start() 
t2.start()

print('Execution complete')
```

**From the above output we see that the main thread performs the task before the individual threads complete execution. In total this takes a time of 10 seconds. But we want the main thread to perform this task at the end. To ensure it does to execute before t1 and t2 completes execution, and then the main thread executes, we use the join method.**

```python
#for doing this, we will import 'thread' class from the 'Threading' module. And let class First and second inherit all
#the properties of the thread class.

from time import sleep
from threading import Thread


class First(Thread):
    def run(self):
        for i in range(10):
            print('Multithreading, Thread1')
            sleep(1)

class Second (Thread): 
    def run(self):
        for i in range(10):
            print('Multithreading, Thread2')
            sleep(1)

#we define threads 't1' and 't2' which are two threads belonging to class 'First' and 'Second' respectively. 

t1=First()
t2=Second()

#Using start method to call the run method

t1.start() 
t2.start()


t1.join()
t2.join()


print('Execution complete')
```

##### run method- entry point of each thread. All codes are written within the run mehod
##### Start method- calls the run method. Until we specify the start method, run method is not called and multithreading does not happen.
##### Join method- join method allows each thread to finish execution,then allows the main thred to execute.

* The main thread(basic unit of CPU) is by default created to execute a certain program. this is 'single Threading'.

* Now to save execution time, we create threads wherein each thread is responsible to execute a specific task or function within a program. So,in addition to the main thread(responsible for the entire program execution) we have multiple threads working on specific tasks to ensure faster execution. This parallel execution of multiple threads along with the main thread is called 'Multithreading'

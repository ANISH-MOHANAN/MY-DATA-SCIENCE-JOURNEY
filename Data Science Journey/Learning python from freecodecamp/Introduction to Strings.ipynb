#### What Are String Concatenation and String Interpolation?


```python
my_name_str_1 = "ANISH"
my_name_str_2 = "MOHANAN"

name_str = my_name_str_1 + ' ' + my_name_str_2
print(name_str)
```

```python
# If you try to concatenate a string with a number, you'll get a TypeError:
name = "ANISH MOHANAN"
age = 32

name_and_age = name +' ' + age
print(name_and_age)
```

you can convert the number into a string with the built-in str() function

```python
name = "ANISH MOHANAN"
age = 32

name_and_age = name +' ' + str(age)
print(name_and_age)
```

You can also use the augmented assignment operator for concatenation. This is represented by a plus and equals sign (+=)

```python
name = "ANISH MOHANAN"
age = 32

name_and_age = name
name_and_age += ' ' + str(age)
print(name_and_age)
```

Python has a category of string called f-strings (short for formatted string literals), which allows you to handle interpolation with a compact and readable syntax.

F-strings start with f (either lowercase or uppercase) before the quotes, and allow you to embed variables or expressions inside replacement fields indicated by curly braces ({})

```python
name = "Anish Mohanan"
age = 32

name_and_age = f"My name is {name} and my age is {age}."
print(name_and_age)
```

```python
num1 = 10
num2 = 90

total = f'The sum of {num1} and {num2} is {num1 + num2}.'
print(total)
```

#### What Is String Slicing and How Does It Work?

String slicing lets you extract a portion of a string or work with only a specific part of it. Here's the basic syntax:

string[start:stop]

```python
my_string = "Hello World"

#If you want to extract characters from a certain index to another, you just separate the start and stop indices with a colon:
print(my_string[0:5]) 

print(my_string[7:]) #This extracts everything from 7 to the last index

print(my_string[:4]) #This extracts everything from 0 to 3 index
```

```python
# You can also omit both the start and stop indices, which will extract the whole string:

my_str = 'Hello world'
print(my_str[:])  
```

Apart from the start and stop indices, there's also an optional step parameter, which is used to specify the increment between each index in the slice.

Here's the syntax for that:

string[start:stop:step]

```python
my_string = "Hello World"
print(my_string[0:11:2])
```

```python
# A helpful trick you can do with the step parameter is to reverse a string by setting step to -1, and leaving start and stop blank:

my_str = 'Hello world'
print(my_str[::-1]) 
```

#### What Are Some Common String Methods?

upper(): Returns a new string with all characters converted to uppercase.

```python
my_string = "Hello world"

upper_case_string = my_string.upper()
print(upper_case_string)
```

lower(): Returns a new string with all characters converted to lowercase.

```python
my_string = "HELLO WORLD"

lower_case_string = my_string.lower()
print(lower_case_string)
```

strip(): Returns a new string with the specified leading and trailing characters removed. If no argument is passed it removes leading and trailing whitespace.

```python
my_str = '  hello world  '

trimmed_str = my_str.strip()
print(trimmed_str)
```

replace(old, new): Returns a new string with all occurrences of old replaced by new

```python
my_str = "Hello World"

replaced_str = my_str.replace('Hello', 'Hi')
print(replaced_str)
```

split(separator): Splits a string on a specified separator into a list of strings. If no separator is specified, it splits on whitespace.

```python
my_str = "Hello World"

seperated_str = my_str.split()
print(seperated_str)
```

join(iterable): Joins elements of an iterable into a string with a separator.

```python
my_list = ['Hello', 'World']

joined_str = ' '.join(my_list)
print(joined_str)
```

startswith(prefix): Returns a boolean indicating if a string starts with the specified prefix.

```python
my_string = "Hello World"

starts_with_hello = my_string.startswith('Hello')
print(starts_with_hello)
```

endswith(suffix): Returns a boolean indicating if a string ends with the specified suffix.

```python
my_str = 'hello world'

ends_with_world = my_str.endswith('world')
print(ends_with_world) 
```

find(substring): Returns the index of the first occurrence of substring, or -1 if it doesn't find one.

```python
my_str= "Hello World"

indexed_str = my_str.find('World')
print(indexed_str)
```

count(substring): Returns the number of times a substring appears in a string.

```python
my_str= "Hello World"

l_count = my_str.count('l')
print(l_count)
```

capitalize(): Returns a new string with the first character capitalized and the other characters lowercased.

```python
my_str = 'hello world'

capitalized_my_str = my_str.capitalize()
print(capitalized_my_str)
```

isupper(): Returns True if all letters in the string are uppercase and False if not.

```python
my_str = 'hello world'

is_all_upper = my_str.isupper()
print(is_all_upper)
```

```python
# islower(): Returns True if all letters in the string are lowercase and False if not.
my_str = 'hello world'

is_all_lower = my_str.islower()
print(is_all_lower) 
```

```python

# title(): Returns a new string with the first letter of each word capitalized.
my_str = 'hello world'

title_case_my_str = my_str.title()
print(title_case_my_str)
```

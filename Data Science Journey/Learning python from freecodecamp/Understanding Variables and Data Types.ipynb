Here are the most common data types you'll use in Python:

Integer: A whole number without decimals, for example, 10 or -5.

```python
integer_var = 25
print("My Integer variable is: ", integer_var)
```

Float: A number with a decimal point, like 4.41 or -0.4.

```python
float_var = 2.5
print("My floating point variable is: ", float_var)
```

String: A sequence of characters enclosed in single or double quotation marks like 'Hello world!'.

```python
string_var = "I will become a  Data Scientist"
print("My Ambition is that:", string_var)
```

Boolean: A true or false type, written as True or False.

```python
boolean_var = True
print("Boolean:", boolean_var)
```

Set: An unordered collection of unique elements, like {4, 2, 0}.

```python
set_var = {2,5,7}
print("My set is:", set_var)
```

Dictionary: A collection of key-value pairs enclosed in curly braces, like {'name': 'John Doe', 'age': 28}.

```python
dictionary_var = {"Name":"Anish", "Age":32}
print("My dictionary is:", dictionary_var)
```

Tuple: An immutable ordered collection, enclosed in brackets, like (7, 8, 4).

```python
tuple_var = (2,5,7)
print("My tuple is:", tuple_var)
```

Range: A sequence of numbers, often used in loops, for example, range(5).

```python
range_var = range(7)
print("My range is:", range_var)
```

List: An ordered collection of elements that supports different data types.

```python
list_var = ["My name is:", "anish","My age is:", 32, True]
print("My list is:", list_var)
```

None: A special value that represents the absence of a value.

```python
none_var = None
print("None variable is:", none_var)
```

To get the data type of a variable, you can use the type() function:

```python
my_var_1 = "Hi"
my_var_2 = 32

print(type(my_var_1))
print(type(my_var_2))
```

```python
my_integer_var = 10
print(type(my_integer_var))  # <class 'int'>

my_float_var = 4.50
print(type(my_float_var))  # <class 'float'>

my_string_var = 'hello'
print(type(my_string_var))  # <class 'str'>

my_boolean_var = True
print(type(my_boolean_var))  # <class 'bool'>

my_set_var = {7, 5, 8}
print(type(my_set_var))  # <class 'set'>

my_dictionary_var = {'name': 'Alice', 'age': 25}
print(type(my_dictionary_var))  # <class 'dict'>

my_tuple_var = (7, 5, 8)
print(type(my_tuple_var))  # <class 'tuple'>

my_range_var = range(5)
print(type(my_range_var))  # <class 'range'>

my_list = [22, 'Hello world', 3.14, True]
print(type(my_list)) # <class 'list'>

my_none_var = None
print(type(my_none_var))  # <class 'NoneType'>
```

The built-in isinstance() function lets you check if a variable matches a specific data type

```python
isinstance("Hello world", str)
```

```python
isinstance(True, bool) # True
```

```python
isinstance(50, int)
```

```python
isinstance("Anish", int)
```

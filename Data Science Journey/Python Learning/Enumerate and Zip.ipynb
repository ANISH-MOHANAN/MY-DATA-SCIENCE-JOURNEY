### enumerate() Tasks

```python
'''
Task 1 – Print Index and Value
fruits = ["Apple", "Banana", "Orange", "Mango"]

Output:

0 Apple
1 Banana
2 Orange
3 Mango
'''

fruits = ["Apple", "Banana", "Orange", "Mango"]

index = 0

for fruit in fruits:
    print(index, fruit)
    index += 1


# using enumerate function
print("\nEnumerate result is")
for index, fruit in enumerate(fruits):
    print(index, fruit)
```

```python
'''
Task 2 – Start Numbering from 1

Using the same list.

Output:

1. Apple
2. Banana
3. Orange
4. Mango
'''

fruits = ["Apple", "Banana", "Orange", "Mango"]

index = 0

for fruit in fruits:
    index += 1
    print(f"{index}. {fruit}")

#using enumerate function
print("\nenumerate function result is")
for index, fruit in enumerate(fruits, start=1):
    print(f"{index}. {fruit}")
```

```python
'''
Task 3 – Find Position
numbers = [25, 40, 15, 60, 90]

Print the position of every number greater than 50.

Expected:

Index 3 : 60
Index 4 : 90
'''

numbers = [25, 40, 15, 60, 90]

index = 0

for number in numbers:
    if number > 50:
        print(f"Index {index} : {number}")
    index += 1


# using enumerate function
print("\nEnumerate function result is")
for index, number in enumerate(numbers):
    if number > 50:
        print(f"Index {index} : {number}")
```

```python
'''
Task 4 – Student Roll Numbers
students = [
    "John",
    "Alice",
    "Bob",
    "Emma"
]

Output:

Roll No 101 : John
Roll No 102 : Alice
Roll No 103 : Bob
Roll No 104 : Emma
'''

students = [
    "John",
    "Alice",
    "Bob",
    "Emma"
]

index = 101

for student in students:
    print(f"Roll No {index} : {student}")
    index += 1


# using enumerate function
print("\nEnumerate function result is")
for index, student in enumerate(students, start=101):
    print(f"Roll No {index} : {student}")
```

### zip() Tasks

```python
'''
Task 5 – Combine Two Lists
names = ["John", "Alice", "Bob"]

marks = [82, 91, 74]

Output:

John : 82
Alice : 91
Bob : 74
'''

names = ["John", "Alice", "Bob"]

marks = [82, 91, 74]

for index in range(len(names)):
    print(f"{names[index]} : {marks[index]}")

# using zip function
print("\nZip function result is")
for name, mark in zip(names, marks):
    print(f"{name} : {mark}")
```

```python
'''
Task 6 – Create Dictionary

Using the same lists,

Create:

{
    "John":82,
    "Alice":91,
    "Bob":74
}
'''

names = ["John", "Alice", "Bob"]

marks = [82, 91, 74]

result = {}

for i in range(len(names)):
     result[names[i]] = marks[i]

print(result)


#using zip function best method
print("\nZip function result is")

result = dict(zip(names, marks))

print(result)

# using dictionary comprehension
print("\n using zip in dictionary comprehension")
result = {name: mark for name, mark in zip(names, marks)}

print(result)
```

```python
'''
Task 7 – Student Grades
students = ["John", "Alice", "Bob", "Emma"]

marks = [82, 91, 68, 97]

Output:

John -> B
Alice -> A
Bob -> D
Emma -> A

Grading:

90+ = A
80-89 = B
70-79 = C
Below 70 = D
'''

students = ["John", "Alice", "Bob", "Emma"]

marks = [82, 91, 68, 97]

for student, mark in zip(students, marks):
    if mark >= 90:
        grade = "A"
    elif mark >= 80:
        grade = "B"
    elif mark >= 70:
        grade = "C"
    else:
        grade = "D"
        
    print(f"{student} -> {grade}")
```

```python
'''
Task 8 – Compare Lists
products = [
    "Laptop",
    "Mouse",
    "Keyboard"
]

prices = [
    65000,
    800,
    1500
]

Print:

Laptop costs ₹65000

Mouse costs ₹800

Keyboard costs ₹1500
'''

products = [
    "Laptop",
    "Mouse",
    "Keyboard"
]

prices = [
    65000,
    800,
    1500
]

for product, price in zip(products, prices):
    print(f"{product} costs ₹{price}")
```

```python
'''
Task 9 – Total Bill
products = [
    "Laptop",
    "Mouse",
    "Keyboard"
]

prices = [
    65000,
    800,
    1500
]

quantity = [
    2,
    5,
    3
]

Using zip() calculate:

Laptop = ₹130000

Mouse = ₹4000

Keyboard = ₹4500

Also calculate:

Total Bill = ?
'''

products = [
    "Laptop",
    "Mouse",
    "Keyboard"
]

prices = [
    65000,
    800,
    1500
]

quantity = [
    2,
    5,
    3
]

total_bill = 0
for product, price, qty in zip(products, prices, quantity):
    bill = price * qty
    total_bill += bill
    print(f"{product} = ₹{bill}")
    
print(f"\nTotal Bill = ₹{total_bill}")
```

```python
'''
Task 10 – Highest Scorer
students = [
    "John",
    "Alice",
    "Bob",
    "Emma"
]

marks = [
    82,
    91,
    68,
    97
]

Find:

Top Student : Emma

Marks : 97

Use zip().

Don't use max().
'''


students = [
    "John",
    "Alice",
    "Bob",
    "Emma"
]

marks = [
    82,
    91,
    68,
    97
]

#initializing variables
top_mark = None
top_student = ""

# starting loop using zip function
for student, mark in zip(students, marks):

    #comparing marks
    if top_mark is None or mark > top_mark:
        
        top_mark = mark
        top_student = student

print(f"Top student : {top_student}")
print(f"Marks : {top_mark}")
```

### Mini Challenge – Employee Report

```python
'''
employees = [
    "John",
    "Alice",
    "Bob",
    "Emma"
]

departments = [
    "HR",
    "IT",
    "Finance",
    "IT"
]

salary = [
    50000,
    65000,
    45000,
    72000
]

Using zip():

Print:

John works in HR and earns ₹50000

Alice works in IT and earns ₹65000

...

Then print only employees earning more than ₹50,000.
'''

employees = [
    "John",
    "Alice",
    "Bob",
    "Emma"
]

departments = [
    "HR",
    "IT",
    "Finance",
    "IT"
]

salaries = [
    50000,
    65000,
    45000,
    72000
]


for employee, department, salary in zip(employees, departments, salaries):
    print(f"{employee} works in {department} and earns ₹{salary}")

print("\nEmployees earning more than ₹50,000:")

for employee, department, salary in zip(employees, departments, salaries):
    if salary > 50000:
        print(f"{employee} works in {department} and earns ₹{salary}")
```

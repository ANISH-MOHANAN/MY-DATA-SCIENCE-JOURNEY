```python
'''
Task 1

Use the math module to print:

Square root of 144
Value of π
5³
'''

import math

square_root = math.sqrt(144)
value_pi = math.pi
cube_value = math.pow(5, 3)

print("Square root of 144 is:", square_root)
print("Value of π is:", value_pi)
print("5³ is:", cube_value)
```

```python
'''
Task 2

Use the random module.

Create a list:

colors = ["Red", "Blue", "Green", "Black"]

Print one random color.
'''

import random

colors = ["Red", "Blue", "Green", "Black"]

random_color = random.choice(colors)

print(random_color)
```

```python
'''
Task 4

Generate 5 random integers between 1 and 100.
'''

import random

for i in range(5):
    print(random.randint(1, 100))
```

```python
'''
Find the largest of the 5 random numbers.
'''

import random

numbers =[]

for _ in range(5):
    number = random.randint(1, 100)
    numbers.append(number)
print("Numbers:", numbers)
print("Largest of 5 random numbers =", max(numbers))
```

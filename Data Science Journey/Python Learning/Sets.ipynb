```python
'''
🟢 Task 1: Remove Duplicates
numbers = [10, 20, 10, 30, 20, 40, 50, 40]

Convert the list into a set.

Expected output:

{10, 20, 30, 40, 50}
'''

numbers = [10, 20, 10, 30, 20, 40, 50, 40]

new_set = set(numbers)

print(new_set)
```

```python
'''
🟢 Task 2: Add an Element
fruits = {"Apple", "Banana", "Orange"}

Add:

Mango

Print the updated set.
'''

fruits = {"Apple", "Banana", "Orange"}

fruits.add('Mango')

print(fruits)
```

```python
'''
🟢 Task 3: Remove an Element

Remove:

Banana

using remove().
'''

fruits = {"Apple", "Banana", "Orange"}

fruits.remove('Banana')

print(fruits)
```

```python
# using discard method because discrad won't raise key error if element is missing

fruits = {"Apple", "Banana", "Orange"}

fruits.discard('Banana')

print(fruits)
```

```python
'''
🟢 Task 4: Membership

Ask the user:

Enter a fruit:

Print:

Available

or

Not Available

using in.
'''


fruits = {"Apple", "Banana", "Orange"}

user_input = input("Enter a fruit: ")

if user_input.lower() in {fruit.lower() for fruit in fruits}:
    print("Available")
else:
    print("Not Available")
```

```python
'''
🟡 Task 5: Union
set1 = {1, 2, 3}
set2 = {3, 4, 5}

Output:

{1, 2, 3, 4, 5}
'''

set1 = {1, 2, 3}
set2 = {3, 4, 5}

set_union = set1.union(set2)

print(set_union)
```

```python
set1 = {1, 2, 3}
set2 = {3, 4, 5}

set_union = set1 | set2

print(set_union)
```

```python
'''
🟡 Task 6: Intersection

Using the same sets:

Output:

{3}
'''

set1 = {1, 2, 3}
set2 = {3, 4, 5}

set_intersection = set1.intersection(set2)
print(set_intersection)
```

```python
set1 = {1, 2, 3}
set2 = {3, 4, 5}

set_intersection = set1 & set2
print(set_intersection)
```

```python
'''
🟡 Task 7: Difference

Output:

{1, 2}
'''
# differnce keeps the ones unique to the first set only discards common 
# one and the ones in the second set
set1 = {1, 2, 3}
set2 = {3, 4, 5}

set_difference = set1.difference(set2)
print(set_difference)
```

```python
set1 = {1, 2, 3}
set2 = {3, 4, 5}

set_difference = set1 - set2
print(set_difference)
```

```python
# symmetric difference is the ones which are not common in both sets

set1 = {1, 2, 3}
set2 = {3, 4, 5}

symmetric_difference = set1 ^ set2
print(symmetric_difference)
```

```python
set1 = {1, 2, 3}
set2 = {3, 4, 5}

symmetric_difference = set1.symmetric_difference(set2)
print(symmetric_difference)
```

```python
'''
🔴 Task 8: Mini Challenge

Given two customer lists:

customers_2024 = {
    "John",
    "Alice",
    "Bob",
    "Emma"
}

customers_2025 = {
    "Bob",
    "Emma",
    "David",
    "Sophia"
}

Find:

Customers who stayed (intersection)
New customers (difference)
Customers who left (difference)
Total unique customers (union)
'''

customers_2024 = {
    "John",
    "Alice",
    "Bob",
    "Emma"
}

customers_2025 = {
    "Bob",
    "Emma",
    "David",
    "Sophia"
}

# intersection
customers_who_stayed = customers_2024 & customers_2025

# difference
new_customers = customers_2025 - customers_2024

# difference
customers_who_left =  customers_2024 - customers_2025

# union
total_unique_customers = customers_2024 | customers_2025

print("Customers who stayed:", customers_who_stayed)
print("New customers:", new_customers)
print("Customers who left:", customers_who_left)
print("Total unique customers:", total_unique_customers)
```

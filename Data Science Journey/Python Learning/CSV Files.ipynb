```python
'''
Task 8 – Count Students by Course

Expected output:

Data Science : 3
AI : 2
Computer Science : 1
Cyber Security : 1
'''

import csv

with open("students.csv", "r", encoding= "utf-8") as csv_file:
    csv_reader = csv.DictReader(csv_file)

    # initializing empty dictionary
    courses = {}
    
    # loop through each student
    for line in csv_reader:

        # getting course name using Course key
        course = line["Course"]
        
        # Add a new course to the dictionary or increment its count if it already exists
        if course not in courses:
            courses[course] = 1
        else:
            courses[course] += 1

    # getting key, value pair from Courses dictionary
    for key, value in courses.items():
        print(f"{key} : {value}")
```

```python
'''
Task 10 – Write a New CSV

Create:

passed_students.csv

Only write students whose marks are 80 or above.

The output file should contain:

Name,Age,Course,Marks
Anish,27,Data Science,92
Alice,22,Computer Science,85
John,23,AI,88
Emma,25,Cyber Security,95
Sophia,21,AI,81
'''

import csv

with open("students.csv", "r", encoding = "utf-8") as csv_file:
    # Create the reader
    csv_reader = csv.DictReader(csv_file)

    # Open the output file
    with open("passed_students.csv", "w", newline="", encoding = "utf-8") as new_file:
        
        # Automatically get headers from the input CSV
        fieldnames = csv_reader.fieldnames
    
        # Create the writer
        csv_writer = csv.DictWriter(new_file, fieldnames= fieldnames)

        # Write the header
        csv_writer.writeheader()
    
        # Read each student
        for line in csv_reader:

            # getting mark and converting to integer
            mark = int(line["Marks"])

            if mark >= 80:
                csv_writer.writerow(line)             
```

```python
'''
Mini Challenge – Student Management System (CSV Version)

Create this menu:

1. Add Student
2. Show Students
3. Search Student
4. Top Student
5. Average Marks
6. Exit

Store all data in:

students.csv

This combines almost everything you've learned so far:

Functions
Loops
Dictionaries
File handling
CSV
Searching
Writing files
'''

import csv
import os

FILE_NAME = "students_record.csv"

FIELDNAMES = ["Name", "Marks"]


# Create the CSV file with headers if it doesn't exist
def ensure_file_exists():

    if not os.path.exists(FILE_NAME):

        with open(FILE_NAME, "w", newline="", encoding="utf-8") as file:

            writer = csv.DictWriter(file, fieldnames=FIELDNAMES)

            writer.writeheader()


def menu():
    print("\n===== Student Management System =====")
    print("\n1. Add Student")
    print("2. Show Students")
    print("3. Search Student")
    print("4. Top Student")
    print("5. Class Average")
    print("6. Exit")




# function to validate marks
def validate_mark():
    
    # starting the loop 
    while True:

        # asking for user input for mark
        mark = input("Enter marks (0-100):")

        # checking whether marks is integer or not
        if not mark.isdigit():
            print("Enter numbers only")
            continue

        mark = int(mark)

        # checking the range of marks. marks must be between 0 -100
        if 0 <= mark <= 100:
            return mark
            
        print("mark must be between 0 - 100")




#function to check whether student already exists
def student_exists(name):

    # calling ensure_file_exists function
    ensure_file_exists()

    name = name.strip().lower()

    # Opening file in read mode to check for duplicates
    with open(FILE_NAME, "r", newline = "", encoding = "utf-8") as file:

        reader = csv.DictReader(file)

        # looping through the file to get details
        for row in reader:
            
            if row["Name"].lower() == name:
                return True   
    
    return False




# function to add student
def add_student():

    # asking for user input for student name
    name = input("Enter student name:").strip()

    # check for duplicates
    if student_exists(name):
        print("student already exists")
        return

    # asking for number of subjects
    subjects = int(input("Enter number of subjects: "))

    # empty list
    marks = []
    
    # cpllecting multiple marks on the basis of no.of subjects
    for i in range(subjects):

        print(f"Subject {i+1}")
        
        mark = validate_mark()

        #storing mark as list
        marks.append(mark) 

    # converting list into one string
    mark_string = ",".join(str(mark) for mark in marks)

    # appending student name and mark to the file
    with open(FILE_NAME, "a", newline = "", encoding = "utf-8") as file:

        writer = csv.DictWriter(file, fieldnames = FIELDNAMES)

        writer.writerow({
            "Name": name,
            "Marks": mark_string
        })

        
    print("\nStudent added successfully.")





# Function to show all students
def show_students():

    # calling ensure_file_exists function
    ensure_file_exists()

    with open(FILE_NAME, "r", encoding="utf-8") as file:

        # reading file contents 
        reader = csv.DictReader(file)

        # storing contents as a list
        students = list(reader)

        # if students is empty
        if not students:
            print("\nNo student records found.")
            return

        print("\nStudent Records")
        print("-" * 30)
        print(f'{"Name":<20}{"Marks":<20}{"Total":<20}{"Average"}')
        print("-" * 30)

        
        for row in students:
            marks_string = row["Marks"].split(",")

            # convrting string values to list of integers
            marks_int = list(int(mark) for mark in marks_string)

            # marks_int = [int(mark) for mark in marks_string]
            # marks_int = list(map(int, marks_string))

            total = sum(marks_int)

            average = total / len(marks_int)

        
        
            print(f'{row["Name"]:<20}{row["Marks"]:<20}{total:<20}{average:.2f}')



def show_marks():

    ensure_file_exists()

    with open(FILE_NAME, "r", newline="", encoding="utf-8") as file:

        reader = csv.DictReader(file)

        print("\nStudent Marks")
        print("-" * 30)

        for row in reader:
            print(f'{row["Name"]:<20}{row["Marks"]}')
            
        

# Function to search student
def search_student():

    # calling ensure_file_exists function
    ensure_file_exists()

    search = input("Enter student name: ").strip().lower()

    found = False

    with open(FILE_NAME, "r", newline = "", encoding="utf-8") as file:

        # reading file contents 
        reader = csv.DictReader(file)
    
            
        # looping through the file to get details
        for row in reader:

            if row["Name"].lower() == search:

                found = True

                # converting marks to string of marks
                marks_string = row["Marks"].split(",")

                # converting strings to integers
                marks_int = [int(mark) for mark in marks_string]
                
                average = sum(marks_int) / len(marks_int)


                print("\nStudent Found")
                print("-" * 20)

                # Print all key-value pairs (generic solution)
                for key, value in row.items():
                    print(f"{key}: {value}")

                print(f"Average: {average:.2f}")

                break


    if not found:
        print("Student Not Found")
        

            


# function to FIND top student based on total marks
def top_student():

    # calling ensure_file_exists function
    ensure_file_exists()

    top_name = ""
    top_total = None

    with open(FILE_NAME, "r", newline = "", encoding = "utf-8") as file:

        # reading file contents 
        reader = csv.DictReader(file)
        
        for row in reader:

            # converting marks to string of marks
            marks_string = row["Marks"].split(",")

            # converting strings to integers
            marks_int = [int(mark) for mark in marks_string]

            total = sum(marks_int)

            # comparing total marks
            if top_total is None or total > top_total:
                top_total = total 
                top_name = row["Name"]


    if top_total is None:
        print("No student records found.")
    else:
        print(f"\nTop Student : {top_name}")
        print(f"Marks       : {top_total}")

                

        
# function to calculatr average marks
def class_average():

    # calling ensure_file_exists function
    ensure_file_exists()

    overall_total = 0
    overall_count = 0

    with open(FILE_NAME, "r", newline = "", encoding = "utf-8") as file:

        reader = csv.DictReader(file)

        for row in reader:

            # converting marks to string of marks
            marks_string = row["Marks"].split(",")

            # converting strings to integers
            marks_int = [int(mark) for mark in marks_string]
                        
            # Add marks to the total
            overall_total += sum(marks_int)
            overall_count += len(marks_int)

    if overall_count == 0:
        return None
        
    # Finding class average
    class_average = overall_total / overall_count
    
    return class_average

      


# MAIN menu loop function
def student_record():

    ensure_file_exists()

    while True:

        menu()

        # asing to enter option from menu and handling error if alphabets are entered
        try:
            option = int(input("\nSelect an option: "))

        except ValueError:
            print("Please enter numbers only.")
            continue

        if option == 1:
            add_student()

        elif option == 2:
            show_students()

        elif option == 3:
            search_student()

        elif option == 4:
            top_student()

        elif option == 5:

            show_marks()
            
            average = class_average()

            if average >= 90:
                print("Excellent class performance!")

            if average is None:
                print("No student records found.")
            else:
                print(f"\nClass Average: {average:.2f}")

        elif option == 6:
            print("Thank you for using student Management System")
            break

        else:
            print("Choose a valid option")

student_record()
```

```python
'''
Bonus Challenge (Highly Recommended)

Create another CSV file:

Product,Category,Price,Quantity
Laptop,Electronics,65000,5
Mouse,Electronics,800,25
Keyboard,Electronics,1500,15
Chair,Furniture,4500,8
Desk,Furniture,7000,4
Monitor,Electronics,12000,6

Find:

Total inventory value (Price × Quantity)
Most expensive product
Cheapest product
Products costing more than ₹5000
Number of products in each category
Average product price
'''

def read_products():

    with open("Products.csv", "r", newline="", encoding="utf-8") as file:

        reader = csv.DictReader(file)

        return list(reader)

products = read_products()



# Function to find Total inventory value (Price × Quantity)
def total_inventory_value(products):
    
        total_inventory = 0

        for row in products:

            # converting price value from string to integer
            price = int(row["Price"])

            # converting quantity value from string to integer
            quantity = int(row["Quantity"])

            inventory_value = price * quantity

            total_inventory += inventory_value

            # printing product wise inventory value
            print(f'{row["Product"]}: ₹{inventory_value}')

        # printing total inventory value
        print(f"Total Inventory Value: ₹{total_inventory}")




# function to find Most expensive product
def most_expensive_product(products):
         
        highest_price = None
        expensive_product = ""

    
        for row in products:
            
            # converting price value from string to integer
            price = int(row["Price"])

            # Compare prices
            if highest_price is None or price > highest_price:
                highest_price = price
                expensive_product = row["Product"]
                
        print("\nMost Expensive Product")
        print("-" * 30)
        print(f"Product: {expensive_product}")
        print(f"Price: ₹{highest_price}")




# function to find cheapest product
def cheapest_product(products):
        
        lowest_price = None
        cheapest_product_name = ""

    
        for row in products:
            
            # converting price value from string to integer
            price = int(row["Price"])

            # Compare prices
            if lowest_price is None or price < lowest_price:
                lowest_price = price
                cheapest_product_name = row["Product"]
                
        print("\nCheapest Product")
        print("-" * 30)
        print(f"Product: {cheapest_product_name}")
        print(f"Price: ₹{lowest_price}")




# function to find Products costing more than ₹5000
def product_cost_5000(products, limit):
 
        print("\nThe products which costs more than 5000 is:")
        
        for row in products:
            
            # converting price value from string to integer
            price = int(row["Price"])

            product = row["Product"]
            
            if price > limit:
                print(product)





#function to find no.of products in each category    
def products_in_category(products):

        # initializing empty categories dictionary
        categories = {}

        # looping through file
        for row in products:

            # getting category
            category = row["Category"]

            # counting products in category 
            if category not in categories:
                categories[category] = 1
            else:
                categories[category] += 1

            # using get method
            #categories[category] = categories.get(category, 0) + 1

        print("\nNumber of products in each category is:")
        
        # getting categories and no.of products as key,value pairs
        for key, value in categories.items():
            print(f"{key}: {value}")





# function to find Average product price
def average_product_price(products):

        total_price = 0
        count = 0

               
        for row in products:

            # converting price from string to integer
            price = int(row["Price"])

            # adding each product price
            total_price += price

            # counting products
            count += 1

        if count == 0:
            print("No products found.")
            return
    
        average = total_price / count
    
        print(f"\nAverage product price is: ₹{average:.2f}")





def product_inventory_management_system(products):

    print("Product Inventory Management System\n")

    # Function to find Total inventory value (Price × Quantity)
    total_inventory_value(products)

    # function to find Most expensive product
    most_expensive_product(products)

    # function to find cheapest product
    cheapest_product(products)

    # function to find Products costing more than ₹5000
    product_cost_5000(products,5000)

    #function to find no.of products in each category    
    products_in_category(products)

    # function to find Average product price
    average_product_price(products)


product_inventory_management_system(products)
```

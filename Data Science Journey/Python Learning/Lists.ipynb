```python
'''
Task 7: Largest Number

You've already solved this before.

Now solve it using a function:

def largest(numbers):

Return the largest number.
'''

numbers = [5, 10, 15, 20]

def largest(numbers):
    largest = None

    for i in numbers:
        if largest is None or i > largest:
            largest = i
    return largest

print("Largest number is:", largest(numbers))
```

```python
'''
Task 8: Smallest Number

Same idea.

Create:

def smallest(numbers):
'''

numbers = [5, 10, 15, 20]

def smallest(numbers):
    smallest = None

    for i in numbers:
        if smallest is None or i < smallest:
            smallest = i
    return smallest

print("Smallest is:", smallest(numbers))
```

```python
'''
Task 9: Count Even Numbers

Create:

def count_even(numbers):

Return the number of even values.
'''

numbers = [5, 10, 15, 20]

def count_even(numbers):
    even_count = 0

    for i in numbers:
        if i % 2 == 0:
            even_count += 1
    return even_count

print("Even number count is:", count_even(numbers))
```

```python
'''
Task 10: Average

Create:

def average(numbers):

Return the average.
'''

numbers = [5, 10, 15, 20]

def average(numbers):
    if not numbers:
        return 0
        
    total = 0
    count = 0

    for i in numbers:
        total += i
        count += 1
    average = total / count
    return average

print("Average is:", average(numbers))
```

```python
'''
Task 13: Membership

Ask the user:

Enter fruit:

If the fruit exists in the list:

Found

Otherwise:

Not Found

Use:

in
'''

fruits = ["Apple", "Mango", "Pineapple", "Orange"]

def membership(fruits):
    fruit = input("Enter fruit:")

    # list comprehension
    if fruit.lower() in [item.lower() for item in fruits]:
        return "Found"
    else:
        return "Not Found"

print(membership(fruits))
```

### Level 4: Interview Questions

```python
'''
Task 14: Second Largest Number
numbers = [10, 40, 25, 90, 60]

Output:

Second largest = 60

Don't use sort() or max().
'''

numbers = [10, 40, 25, 90, 60]

def second_largest(numbers):
    largest = None
    second_largest = None

    for i in numbers:
        if largest is None or i > largest:
            second_largest = largest
            largest = i
        elif (second_largest is None or i > second_largest) and i != largest:
            second_largest = i
    return second_largest

print("Second largest =", second_largest(numbers))
```

```python
'''
ask 15: Remove Duplicates
numbers = [1, 2, 2, 3, 4, 4, 5]

Output:

[1, 2, 3, 4, 5]
'''

numbers = [1, 2, 2, 3, 4, 4, 5]

def remove_duplicate(numbers):
    new_list = []
    
    for i in numbers:
        if i not in new_list:
            new_list.append(i)
    return new_list

print(remove_duplicate(numbers))
```

```python
'''
Task 16: Count Frequency
numbers = [1, 2, 2, 3, 2, 4, 5]

Output:

2 appears 3 times
'''

numbers = [1, 2, 2, 3, 2, 4, 5]

def frequency(numbers):
    count = 0
    
    for i in numbers:
        if i == 2:
            count += 1
    return count

print("2 appears", frequency(numbers), "times")
```

```python
numbers = [1, 2, 2, 3, 2, 4, 5]

def frequency(numbers):
    checked = []

    for i in numbers:
        if i not in checked:
            count = 0

            for j in numbers:
                if i == j:
                    count += 1
            if count > 1:
                return i, count

            checked.append(i)

number, count = frequency(numbers)
print(number, "appears", count, "times")
```

```python
'''
Project 1: Student Marks

Create:

marks = []

Ask the user to enter 5 marks.

After that print:

Highest mark
Lowest mark
Average
Pass count (>40)
'''



def student_marks():
    
    #creating empty list
    marks =[]

    # asking for user input
    for i in range(5):
        mark = int(input(f"Enter mark {i+1}: "))
        marks.append(mark)
        
    # Declaring variables
    highest_mark = None
    lowest_mark = None
    pass_count = 0
    total = 0
    count = 0
    
    # Looping to find highest mark, lowest mark, and pass count
    for mark in marks:
        if highest_mark is None or mark > highest_mark:
            highest_mark = mark
        if lowest_mark is None or mark < lowest_mark:
            lowest_mark = mark
        if mark > 40:
            pass_count += 1
            
        # finding total and count    
        total += mark
        count += 1
        
    # Finding average    
    average = total / count
    
    return highest_mark, lowest_mark, average, pass_count

# Unpacking
(highest, lowest, average, passed) = student_marks()
    
print("Highest mark:", highest) 
print("Lowest mark:", lowest)
print("Average:", average)
print("Pass count:", passed)   
```

```python
'''
Project 2: Shopping Cart

Create:

cart = []

Menu:

1. Add item
2. Remove item
3. Show cart
4. Exit

Keep running until the user chooses Exit.
'''

def shopping_cart():

    #empty list
    cart = []

    # print menu
    def menu():
        print("1. Add item")
        print("2. Remove item")
        print("3. Show cart")
        print("4. Exit")
    
    # starting loop
    while True:
        
        # showing menu
        menu()

        # asing to enter option from menu and handling error if alphabets are entered
        try:
            option = int(input("Choose the menu option:"))
        except ValueError:
            print("Please enter a number.")
            continue
            
        # adding item to the cart
        if option == 1:
            item = input("Enter item to add to the cart:")
            cart.append(item)
            print(f"{item} added to the cart.")
            

        # removing item
        elif option == 2:
            item = input("Enter item to remove from the cart:")
            if item in cart:
                cart.remove(item)
                print(f"{item} removed from the cart.")
            else:
                print("Item not found in the cart.")

        # showing items in the cart
        elif option == 3:
            print(cart)

        # exiting
        elif option == 4:
            print("Thank you for shopping.")
            break
            
        # print invalid option if option entered is not from 1 to 4 
        else:
            print("Invalid option")

# function calling
shopping_cart()
```

```python
'''
Project 3: To-Do List

Menu:

1. Add Task
2. Remove Task
3. Show Tasks
4. Exit
'''

def to_do_list():

    # empty list
    to_do = []

    # print menu
    def menu():
        print("\n1. Add Task")
        print("2. Remove Task")
        print("3. Show Tasks")
        print("4. Exit")

    # starting loop
    while True:

        # showing menu
        menu()

        #asing to enter option from menu and handling error if alphabets are entered
        try:
            option = int(input("Choose a menu option:"))
        except ValueError:
            print("Please enter a valid number")
            continue

        # adding tasks to the to do list
        if option == 1:
            task = input("Enter the task to add to the list:")
            if task not in to_do:
                to_do.append(task)
                print(f"{task} added to the To-Do List")
            else:
                print(f"{task} is already in the To-Do List")

        # Removing task from the list
        elif option == 2:
            task = input("Enter the task to be removed from the To do list:")
            if task in to_do:
                to_do.remove(task)
                print(f"{task} removed from the To-Do List")
            else:
                print("Task not found in the To-Do List.")

        # showing the list
        elif option == 3:
            if not to_do:
                print("Your To-Do List is empty.")
            else:
                print("Your To-Do List is:")
                
                # enumerate function and for loop to print the tasks as a list
                for number, task in enumerate(to_do, start = 1):
                    print(f"{number}. {task}")

        # Exiting the list
        elif option == 4:
            print("Thank you for using the To-Do List.")
            break

        # print invalid option if option entered is not from 1 to 4 
        else:
            print("Invalid option")

to_do_list()
```

```python
'''
Mini Contact Book
1. Add Contact
2. Delete Contact
3. Search Contact
4. Show Contacts
5. Exit
'''

def contact_book():

    # empty contact book
    book = []

    # function to print menu
    def menu():
        print("\n1. Add Contact")
        print("2. Delete Contact")
        print("3. Search Contact")
        print("4. Show Contacts")
        print("5. Exit")

    # starting loop
    while True:

        # showing menu
        menu()

        #asing to enter option from menu and handling error if alphabets are entered
        try:
            option = int(input("\nChoose a menu option:"))
        except ValueError:
            print("\nPlease enter a valid number")
            continue

        # Add Contact
        if option == 1:
            contact = input("\nEnter the contact:").lower()
            
            if contact not in book:
                book.append(contact)
                print(f"{contact} added to the contact list")
            else:
                print(f"{contact} already in the contact list")

        # Delete Contact        
        elif option == 2:
            contact = input("\nEnter the contact to be deleted:").lower()
            
            if contact in book:
                book.remove(contact)
                print(f"{contact} removed from the contact list")
            else:
                print(f"{contact} not found in the list")

        # Search Contact
        elif option == 3:
            contact = input("\nEnter the contact to search:").lower()
            
            if contact in book:
                print(f"{contact} found")
            else:
                print(f"{contact} not found")

        # Show Contacts
        elif option == 4:
            if not book:
                print("\nYour contact list is empty")
            else:
                print("\nYour conatct list is:")

                # enumerate function and for loop to print the tasks as a list
                for number, contact in enumerate(book, start = 1):
                    print(f"{number}. {contact}")

        # exiting
        elif option == 5:
            print("\nThank you for using the contact book")
            break

        # handling error if option selected is beyond 5
        else:
            print("\nInvalid option")

contact_book()       
```

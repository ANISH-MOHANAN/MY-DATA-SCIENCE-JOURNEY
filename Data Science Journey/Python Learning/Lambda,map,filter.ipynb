### Lambda Tasks

```python
'''
Task 1 – Square

Create a lambda function that returns the square of a number.

Example:

Input: 6

Output:
36
'''

def square(number):
    return number * number
print(square(2))

# lambda function
square = lambda number: number * number
print("Lambda function result")
print(square(6))
```

```python
'''
Task 2 – Cube

Create a lambda function that returns the cube of a number.

Input: 4

Output:
64
'''

def cube(number):
    return number ** 3

print(cube(4))


# lambda function
cube = lambda number: number ** 3
print("Lambda function result")
print(cube(4))
```

```python
'''
Task 3 – Even or Odd

Create a lambda function that returns:

"Even"

or

"Odd"

depending on the number.

Example:

8 → Even
15 → Odd
'''

def even_odd(number):
    if number % 2 == 0:
        return "Even"
    else:
        return "Odd"
print(even_odd(8))


# lambda function
even_odd = lambda number: "Even" if number % 2 == 0 else "Odd"
print("\nLambda function result")
print(even_odd(15))
```

```python
'''
Task 4 – Largest Number

Create a lambda function that accepts two numbers and returns the larger one.

Example:

10, 25

Output:
25
'''

def largest(x, y):
    if x > y:
        return x
    else:
        return y
print("Largest is:",largest(10, 25))


# Lambda function 
largest = lambda x,y : x if x > y else y
print("\nLambda function result is:", largest(10, 25))
```

### map() Tasks

```python
'''
Task 5 – Double Every Number
numbers = [2, 4, 6, 8, 10]

Output:

[4, 8, 12, 16, 20]

Use map().
'''

numbers = [2, 4, 6, 8, 10]

def double(number):
   return number * 2
    

# map function
numbers = [2, 4, 6, 8, 10]

double_numbers = map(double, numbers)
print(list(double_numbers))
```

```python
numbers = [2, 4, 6, 8, 10]

double_numbers = map(lambda number: number * 2, numbers)

print(list(double_numbers))
```

```python
'''
Task 6 – Convert to Uppercase
names = ["anish", "alice", "bob", "john"]

Output:

["ANISH", "ALICE", "BOB", "JOHN"]
'''

names = ["anish", "alice", "bob", "john"]

def upper_case(name):
    return name.upper()

uppercase = map(upper_case, names) 
print(list(uppercase))
```

```python
# Using a lambda
names = ["anish", "alice", "bob", "john"]

uppercase = map(lambda name: name.upper(), names)
print(list(uppercase))
```

```python
'''
Task 7 – Add ₹5000 Bonus
salaries = [25000, 32000, 41000, 58000]

Output:

[30000, 37000, 46000, 63000]
'''

salaries = [25000, 32000, 41000, 58000]

def bonus(salary):
    return salary + 5000

salary_with_bonus = map(bonus, salaries)
print(list(salary_with_bonus))
```

```python
'''
Task 8 – Celsius to Fahrenheit
temps = [0, 20, 30, 40]

Formula:

F = (C × 9/5) + 32

Expected:

[32.0, 68.0, 86.0, 104.0]
'''

temps = [0, 20, 30, 40]

def c_to_f(temp):
    return temp * 9/5 + 32

temp_fahrenheit = map(c_to_f, temps)
print(list(temp_fahrenheit))
```

```python
# Using a lambda
temps = [0, 20, 30, 40]

temp_fahrenheit = map(lambda temp: temp * 9/5 + 32, temps)
print(list(temp_fahrenheit))
```

### filter() Tasks

```python
'''
Task 9 – Even Numbers
numbers = [3, 8, 12, 17, 24, 31, 42]

Output:

[8, 12, 24, 42]
'''

numbers = [3, 8, 12, 17, 24, 31, 42]

def is_even(number):
    return number % 2 == 0

even_numbers = filter(is_even, numbers)
print(list(even_numbers))
```

```python
# using lambda

numbers = [3, 8, 12, 17, 24, 31, 42]

even_numbers = filter(lambda number: number % 2 == 0, numbers)
print(list(even_numbers))
```

```python
'''
Task 10 – Numbers Greater Than 50
numbers = [20, 55, 48, 72, 81, 39, 100]

Output:

[55, 72, 81, 100]
'''

numbers = [20, 55, 48, 72, 81, 39, 100]

def is_above_50(number):
    return number > 50

numbers_above_50 = filter(is_above_50, numbers)
print(list(numbers_above_50))
```

```python
# using lambda
numbers = [20, 55, 48, 72, 81, 39, 100]

numbers_above_50 = filter(lambda number: number > 50, numbers)
print(list(numbers_above_50))
```

```python
'''
Task 11 – Long Words
words = [
    "Python",
    "AI",
    "Machine",
    "SQL",
    "Statistics",
    "ML"
]

Return only words whose length is greater than 5.

Expected:

["Python", "Machine", "Statistics"]
'''

words = [
    "Python",
    "AI",
    "Machine",
    "SQL",
    "Statistics",
    "ML"
]

def is_length_above_5(word):
    return len(word) > 5

long_words = filter(is_length_above_5, words)
print(list(long_words))
```

```python
# using lambda
words = [
    "Python",
    "AI",
    "Machine",
    "SQL",
    "Statistics",
    "ML"
]

long_words = filter(lambda word: len(word) > 5, words)
print(list(long_words))
```

### Mini Challenge – Employee Data

```python
'''
Mini Challenge – Employee Data
employees = [
    {"name": "John", "salary": 50000},
    {"name": "Alice", "salary": 65000},
    {"name": "Bob", "salary": 45000},
    {"name": "Emma", "salary": 70000}
]

Complete the following:

Part 1 (map)

Create a list of employee names in uppercase.

Expected:

["JOHN", "ALICE", "BOB", "EMMA"]
Part 2 (filter)

Return only employees earning ₹50,000 or more.

Expected:

[
    {"name": "John", "salary": 50000},
    {"name": "Alice", "salary": 65000},
    {"name": "Emma", "salary": 70000}
]
Part 3 (map)

Increase every employee's salary by 10%.

Return a new list containing the updated salaries.
'''



employees = [
    {"name": "John", "salary": 50000},
    {"name": "Alice", "salary": 65000},
    {"name": "Bob", "salary": 45000},
    {"name": "Emma", "salary": 70000}
]

# function to convert name to upper case
def to_upper(employee):
    return employee["name"].upper()

# map function to create list of employee names in uppercase.
name_in_uppercase = map(to_upper, employees)
print("Employees name in upper case:",list(name_in_uppercase))


# Return only employees earning ₹50,000 or more.
def above_50000(employee):
    return employee["salary"] >= 50000

salary_above_or_equal_50000 = filter(above_50000, employees)

print("\n Emplyees salary more than 50000\n",list(salary_above_or_equal_50000))



# Increase every employee's salary by 10%.
def salary_increase(employee):
    return {
        "name": employee["name"],
        "new_salary": round(employee["salary"] * 1.10, 2)
    }
    
increased_salary = map(salary_increase, employees)
print("\nSalary increased by 10% is:\n", list(increased_salary))
```

```python
# lambda version

# map function to create list of employee names in uppercase.
name_in_uppercase = map(lambda employee: employee["name"].upper(), employees)

print("employee names in uppercase",list(name_in_uppercase))


# Return only employees earning ₹50,000 or more.
salary_above_50000 = filter(
    lambda employee: employee["salary"] >= 50000,
    employees
)

print("\nemployees earning ₹50,000 or more\n", list(salary_above_50000))


# Increase every employee's salary by 10%.
increased_salary = map(lambda employee: employee["salary"] * 1.10, employees)
print("\nIncrease every employee's salary by 10%",list(increased_salary))
```

### Bonus Challenge

```python
'''
students = [
    ("John", 82),
    ("Alice", 91),
    ("Bob", 68),
    ("Emma", 95),
    ("David", 73)
]

Using map() and filter():

Create a list containing only student names.
Create a list of marks increased by 5 (maximum 100).
Return only students scoring 80 or above.
Convert all student names to uppercase.
Create a list of "Name - Grade" where:
A: 90+
B: 80–89
C: 70–79
D: Below 70
'''

students = [
    ("John", 82),
    ("Alice", 91),
    ("Bob", 68),
    ("Emma", 95),
    ("David", 73)
]

# Create a list containing only student names.
def names(student):
    name, mark = student
    return name

student_names = map(names, students)
print("list containing only student names")
print(list(student_names))



# Create a list of marks increased by 5 (maximum 100).
def increase_marks(student):
    name, mark = student
    mark = min(mark + 5, 100)
    return mark

marks_increased = map(increase_marks, students)
print("\nlist of marks increased by 5 (maximum 100)")
print(list(marks_increased))


# Return only students scoring 80 or above.
above_80 = filter(lambda student: student[1] >= 80, students)
print("\nonly students scoring 80 or above.")
print(list(above_80))


# Convert all student names to uppercase.
names_to_uppercase = map(lambda student: student[0].upper(), students)
print("\nstudent names to uppercase")
print(list(names_to_uppercase))



# Create a list of "Name - Grade" where:
# A: 90+
# B: 80–89
# C: 70–79
# D: Below 70


def grades(student):
    name, mark = student
    if mark >= 90:
        return f"{name} - A"
    elif mark >= 80:
        return f"{name} - B"
    elif mark >= 70:
        return f"{name} - C"
    else:
        return f"{name} - D"

name_grade = map(grades, students)
print('\nlist of "Name - Grade"')
print(list(name_grade))
```

```python
students = [
    ("John", 82),
    ("Alice", 91),
    ("Bob", 68),
    ("Emma", 95),
    ("David", 73)
]

# Create a list of marks increased by 5 (maximum 100).
marks_increased = list(
    map(lambda student: min(student[1] + 5, 100), students)
)

print(marks_increased)



# Create a list of "Name - Grade" where:
# A: 90+
# B: 80–89
# C: 70–79
# D: Below 70

name_grade = map(
    lambda student:
        f"{student[0]} - A" if student[1] >= 90 else
        f"{student[0]} - B" if student[1] >= 80 else
        f"{student[0]} - C" if student[1] >= 70 else
        f"{student[0]} - D",
    students
)

print(list(name_grade))
```

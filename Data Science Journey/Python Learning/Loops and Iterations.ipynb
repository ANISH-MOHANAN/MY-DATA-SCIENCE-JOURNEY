### Challenge 3 (Real Interview Style)

```python
'''
Create a number guessing game.

secret = 7

Keep asking user to guess.

Output:

Too high

or

Too low

until correct.

Then print:

Correct!
'''

secret = 7
guess = int(input("Enter your guess:"))

while guess != secret:
    if guess > secret:
        print("Too high")
    elif guess < secret:
        print("Too Low")

    guess = int(input("Enter your guess:"))
    
print("Correct!")
```

### Challenge 3: Average of Numbers

```python
'''
numbers = [10, 20, 30, 40, 50]

Output:

Average = 30

Don't use any built-in average function.
'''

numbers = [10, 20, 30, 40, 50]
total = 0
count = 0

for i in numbers:
    total = total + i
    count = count + 1
    
average = total / count
print("Average =", average)
```

```python
'''
Convert your solution into a function:

def average(numbers):
    # your code here

Then test:

numbers = [10, 20, 30, 40, 50]

print(average(numbers))

Expected output:

30.0
'''

numbers = [10, 20, 30, 40, 50]

def average(numbers):
    total = 0
    
    for i in numbers:
        total += i
        
    count = len(numbers)

    average = total / count
    
    return average

print(average(numbers))
```

### Next Stage: Functions + Loops Together

```python
'''
Task 1: Function to Count Even Numbers

Create:

def count_even(numbers):

Example:

numbers = [12, 45, 7, 89, 23, 67, 20, 18]

print(count_even(numbers))

Output:

3
'''
numbers = [12, 45, 7, 89, 23, 67, 20, 18]

def count_even(numbers):
    count = 0
    
    for i in numbers:
        if i % 2 == 0:
            count = count + 1
    return count
print(count_even(numbers))
```

```python
'''
Task 2: Function to Find Largest Number

Create:

def largest_number(numbers):

Example:

numbers = [12, 45, 7, 89, 23, 67]

print(largest_number(numbers))

Output:

89
'''

numbers = [12, 45, 7, 89, 23, 67]

def largest_number(numbers):
    largest = None

    for i in numbers:
        if largest is None or i > largest:
            largest = i
    return largest

print(largest_number(numbers))
```

```python
'''
Task 3: Function to Count Vowels

Create:

def count_vowels(word):

Example:

print(count_vowels("anish"))

Output:

2
'''

vowels = ["a", "e", "i", "o", "u"]

def count_vowels(word):
    count = 0

    for letter in word.lower():
        if letter in vowels:
            count = count + 1
    return count

print(count_vowels("anish"))
```

```python
'''
Task 4: Function to Calculate Average

Create:

def average(numbers):

Example:

numbers = [10, 20, 30, 40, 50]

print(average(numbers))

Output:

30
'''

numbers = [10, 20, 30, 40, 50]

def average(numbers):
    total = 0
    count = len(numbers)

    for i in numbers:
        total += i
    return total / count
print(average(numbers))
```

```python
'''
Task 5: Function + Condition

Create:

def is_even(number):

Return:

True

or

False

Example:

print(is_even(10))

Output:

True
'''

def is_even(number):
    if number % 2 == 0:
        return True
    else:
        return False

print(is_even(10))
```

### Mini Project 2 (Very Important)

```python
'''
Create a calculator:

def calculator(a, b, operation):

Examples:

calculator(10, 5, "+")

Output:

15

Support:

+
-
*
/
'''

def calculator(a, b, operation):
    if operation == "+":
        return a + b
    elif operation == "-":
        return a - b
    elif operation == "*":
        return a * b
    elif operation == "/":
        if b == 0:
            return "Cannot divide by zero"
        return a / b
    else:
        print("Enter valid numbers or operators")
        return

print(calculator(10, 5, "+"))
```

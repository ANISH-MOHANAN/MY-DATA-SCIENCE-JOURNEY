```python
'''
Task 5: Find Longest Line

File:

Hello
Data Science
Python

Output:

Longest line: Data Science
'''

with open("sample.txt", "w", encoding = "utf-8") as file:
    file.write("Hello\n")
    file.write("Data Science\n")
    file.write("Python")

with open("sample.txt") as file:
    longest = None

    for line in file:
        line = line.strip()
        
        if longest is None:
            longest = line
        elif len(line) > len(longest):
            longest = line

print("Longest line:", longest)
```

### Important Challenge

```python
'''
Count vowels in an entire file.

File:

Hello
Python

Output:

Total vowels = 3
'''

with open("vowels.txt", "w", encoding = "utf-8") as file:
    file.write("Hello\n")
    file.write("Python")

vowels = "aeiou"

with open("vowels.txt") as file:
    count = 0
    
    for line in file:
        for letter in line:
            if letter in vowels:
                count += 1

print("Total vowels =", count)
```

### Most Useful Exercise

```python
'''
Create:

def file_statistics(filename):

Return:

Number of lines
Number of words
Number of characters

Example output:

Lines: 4
Words: 15
Characters: 82

This combines:

Functions
Loops
Strings
File handling

It's one of the best beginner projects.
'''

with open("sample.txt", "w", encoding = "utf-8") as file:
    file.write("Hello Python\n")
    file.write("\n")
    file.write("My name is Anish.\n")
    file.write("I am learning Python.\n")
    file.write("Python is fun to learn.\n")
    file.write("Data Science is my goal.\n")
    file.write("I will become a Data Scientist.\n")

def file_statistics(filename):
    
    with open(filename, "r", encoding = "utf-8") as file:
    
        line_count = 0
        word_count = 0
        letter_count = 0
        
        vowels = "aeiou"
        vowels_count = 0
        consonants_count = 0

        longest_line = ""
        max_length = 0
        
        for line in file:
            
            line = line.strip()

            # line count
            line_count += 1

            #word count
            word_count += len(line.split())

            # check for longest line
            if len(line) > max_length:
                max_length = len(line)
                longest_line = line  

            # count letters, vowels, and consonants
            for letter in line.lower():
                
                if letter.isalpha():
                    
                    letter_count += 1
                
                    if letter in vowels:
                        vowels_count += 1
                    else:
                        consonants_count += 1
                        
        average_words = word_count / line_count  
        
    return (
        line_count, 
        word_count, 
        letter_count, 
        vowels_count, 
        consonants_count,
        longest_line,
        average_words
    )

#unpacking
(
    lines, 
    words, 
    characters, 
    vowels, 
    consonants, 
    longest, 
    average
) = file_statistics("sample.txt")            

print("Lines:", lines)
print("Words:", words)
print("Characters:", characters)
print("Vowels:", vowels)
print("Consonants:", consonants)
print("Longest line:", longest)
print("Average words per line:", round(average, 2))
```

#### Convert temperature Celsius → Fahrenheit

```python
temp_in_celsius = int(input("enter the temperature in celsius: "))
temp_in_farenheit = (temp_in_celsius * 1.8) + 32

print("Temperature in farenheit is: ", temp_in_farenheit)
```

#### Convert Fahrenheit  → temperature Celsius

```python
temp_in_farenheit = int(input("enter the temperature in farenheit: "))
temp_in_celsius = (temp_in_farenheit - 32) / 1.8 

print("Temperature in celsius is: ", temp_in_celsius)
```

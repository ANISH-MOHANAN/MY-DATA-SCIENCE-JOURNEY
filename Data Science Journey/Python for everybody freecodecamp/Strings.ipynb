### Strings Practice 

```python
'''
Task 1: Count Characters

Ask user for a word.

Example:

Enter a word: python

Output:

Length = 6

Hint:

len(word)
'''

word = input("Enter a word:")

print("Length =", len(word))
```

```python
'''
Task 2: Print Each Character

Input:

anish

Output:

a
n
i
s
h

Use a loop.
'''
word = "anish"

for letter in word:
    print(letter)
```

```python
'''
Task 3: Count Vowels

Input:

anish

Output:

Vowels = 2

You said you've already done this. Good.
'''

word = input("Enter the word:").lower()
vowels = ["a", "e", "i", "o", "u"]

def vowels_count(word):
    count = 0

    for letter in word:
        if letter in vowels:
            count += 1
    return count

print("Vowels =", vowels_count(word))
```

```python
'''
Task 4: Count Consonants

Input:

anish

Output:

Consonants = 3
'''

word = input("Enter the word:").lower()
vowels = ["a", "e", "i", "o", "u"]

def consonants_count(word):
    count = 0

    for letter in word:
        if letter.isalpha() and letter not in vowels:
            count += 1
    return count

print("Consonants =", consonants_count(word))
```

### Strings + Loops

```python
'''
Task 5: Reverse a Word

Input:

anish

Output:

hsina

Don't use:

[::-1]

Use a loop.
'''

word = input("Enter the word")
reverse = ""

for letter in word:
    reverse = letter + reverse

print(reverse)
```

```python
'''
Task 6: Count a Specific Letter

Input:

Enter word: banana
Enter letter: a

Output:

Occurrences = 3
'''

word = input("Enter word:").lower()
letter = input("Enter letter:").lower()

def letter_count(word, letter):
    occurrences = 0

    for i in word:
        if i == letter:
            occurrences += 1
    return occurrences

print("Occurrences =", letter_count(word, letter))
            
```

```python
'''
Task 7: Check Palindrome

Input:

madam

Output:

Palindrome

Input:

python

Output:

Not a palindrome
'''

word = input("Enter word:")

def palindrome(word):
    reverse = ""

    for letter in word:
        reverse = letter + reverse

    if word == reverse:
        return 'Palindrome'
    else:
        return 'Not a palindrome'

print(palindrome(word))
```

### Strings + Functions

```python
'''
Task 8

Create:

def count_vowels(word):

Return vowel count.

Example:

print(count_vowels("anish"))

Output:

2
'''

vowels = "aeiou"

def count_vowels(word):
    count = 0

    for letter in word.lower():
        if letter in vowels:
            count += 1
    return count

print(count_vowels("anish"))
```

```python
'''
Task 9

Create:

def reverse_word(word):

Example:

print(reverse_word("python"))

Output:

nohtyp
'''

def reverse_word(word):
    reverse = ""

    for letter in word:
        reverse = letter + reverse
    return reverse

print(reverse_word("python"))
```

```python
'''
Task 10

Create:

def count_letter(word, letter):

Example:

print(count_letter("banana", "a"))

Output:

3
'''

def count_letter(word, letter):
    count = 0

    for i in word:
        if i == letter:
            count += 1
    return count

print(count_letter("banana", "a"))
```

### Real Beginner Challenge

```python
'''
Create:

def password_strength(password):

Rules:

If length < 8

Output:

Weak

Else:

Strong

Example:

password_strength("python")

Output:

Weak

Example:

password_strength("python123")

Output:

Strong
'''

def password_strength(password):     
    if len(password) < 8:
        return "Weak"
    else:
        return "Strong"
print(password_strength("python"))
```

### Most Important Challenge

```python
'''
Create:

def username_checker(username):

Rules:

Length must be at least 5 characters

Output:

Valid Username

or

Invalid Username

Examples:

username_checker("anish")

Output:

Valid Username
username_checker("abc")

Output:

Invalid Username
'''

def username_checker(username):
    if len(username) >= 5:
        return "Valid Username"
    else:
        return "Invalid Username"

print(username_checker("anish"))
print(username_checker("abc"))
```

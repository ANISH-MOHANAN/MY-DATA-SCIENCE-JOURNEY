### Level 3: Functions

```python
'''
Task 11

Create:

def student_details(student):

Print all keys and values.
'''

student = {
    "name": "Anish",
    "age": 28,
    "course": "Data Science"
}

def student_details(student):

    for key, value in student.items():
        print(f"{key} : {value}")

student_details(student)
```

```python
def student_details(student):
    result = ""
    
    for key, value in student.items():
        result += f"{key} : {value}\n"
    return result

print(student_details(student))
```

```python
'''
Task 12

Create:

def average_marks(marks):

Return the average.
'''

def average_marks(marks):

    total = 0

    for value in marks.values():
        total += value
        
    average = total / len(marks)
    return average

print(average_marks(marks))
```

```python
def average_marks(marks):
    return sum(marks.values()) / len(marks)

print(average_marks(marks))
```

```python
'''
Task 13

Create:

def highest_mark(marks):

Return:

Science : 90

Don't use max().
'''

def highest_mark(marks):

    highest_subject = None
    highest_value = None
    
    for subject, value in marks.items():
        if highest_value is None or value > highest_value:
            highest_value = value
            highest_subject = subject
            
    return highest_subject, highest_value

subject, value = highest_mark(marks)

print(f"{subject} : {value}")
```

```python
# Better version to avoid None checks in numbers
def highest_mark(marks):
    highest_subject = None
    highest_value = float('-inf')

    for subject, value in marks.items():
        if value > highest_value:
            highest_value = value
            highest_subject = subject

    return highest_subject, highest_value

subject, value = highest_mark(marks)

print(f"{subject} : {value}")
```

### Real Data Science Style

```python
'''
Task 14: Employee Database
employees = [
    {"name": "John", "salary": 50000},
    {"name": "Alice", "salary": 65000},
    {"name": "Bob", "salary": 45000}
]

Print:

John earns 50000
Alice earns 65000
Bob earns 45000
'''

employees = [
    {"name": "John", "salary": 50000},
    {"name": "Alice", "salary": 65000},
    {"name": "Bob", "salary": 45000}
]

def employee_data(employees):

    for employee in employees:
        print(f"{employee['name']} earns {employee['salary']}")

employee_data(employees)
```

```python
# cleaner version

for employee in employees:
    name = employee["name"]
    salary = employee["salary"]
    print(f"{name} earns {salary}")
```

```python
'''
Task 15: Find Highest Salary

Using the same employees list, print:

Highest Salary
Alice : 65000

Don't use max().
'''

employees = [
    {"name": "John", "salary": 50000},
    {"name": "Alice", "salary": 65000},
    {"name": "Bob", "salary": 45000}
]

def top_salary(employees):

    highest_salary = None
    employee_name = None

    for employee in employees:
        name = employee["name"]
        salary = employee["salary"]

        if highest_salary is None or salary > highest_salary:
            highest_salary = salary
            employee_name = name

    return employee_name, highest_salary


name, salary = top_salary(employees)

print("Highest salary")
print(f"{name} : {salary}")
```

```python
'''
Mini Challenge

Build a simple phone book:

contacts = {}

Menu:

1. Add Contact
2. Search Contact
3. Delete Contact
4. Show Contacts
5. Exit
'''
    
def phone_book():

    # initializing empty dictionary
    contacts = {}

    # function to print menu
    def menu():
        print("\n1. Add Contact")
        print("2. Search Contact")
        print("3. Delete Contact")
        print("4. Show Contacts")
        print("5. Exit")
    
    # starting loop
    while True:
        
        # displaying menu
        menu()
    
        # asing to enter option from menu and handling error if alphabets are entered
        try:
            option = int(input("\nSelect any option from the menu:"))
        except ValueError:
            print("\n Enter a valid option")
            continue

        # add contact
        if option == 1:
            name = input("\nEnter contact name:").lower()

            if not name:
                print("Name cannot be empty.")
                continue

            # if name is not in the contacts dictionary
            if name not in contacts:
                number =input("Enter the mobile number:") 

                number = number.strip()

                # Number format validation
                if number.startswith("+91"):
                    formatted_number = number
                elif number.startswith("91"):
                    formatted_number = "+" + number
                # adding +91 if number entered is not with +91
                else:
                    formatted_number = "+91" + number

                # Handling error if number after + is not numbers 
                # or total length is more than 13
                if ( 
                    not formatted_number[1:].isdigit()
                    or len(formatted_number) != 13
                ):
                    print("Invalid phone number")
                    continue

                # adding name and number to the contacts dictionary if all conditions are satisfied
                contacts[name] = formatted_number
                print(f"\n{name} added to the phone book")

            else:
                print(f"\n{name} already in the phone book")


        # Search Contact
        elif option == 2:
            name = input("\nEnter the contact name to search:").lower()

            if name in contacts:
                print(f"\nContact Found")
                number = contacts[name]
                print(f"\n{name} : {number}")
            else:
                print(f"\n{name} Not Found")

            
        # Delete Contact
        elif option ==3:
            name = input("\nEnter the contact name to delete:").lower()

            if name in contacts:
                del contacts[name]
                print(f"\n{name} deleted")
            else:
                print(f"\n{name} not found")


        # Show Contacts
        elif option == 4:
            if contacts:

                count = 1
                for name, number in contacts.items():
                    print(f"{count}. {name} : {number}")
                    count += 1
            else:
                print("\nPhone book is empty")


        # Exit
        elif option == 5:
            print("\nThank you for using the phone book")
            break
            
phone_book()             
```

```python
'''
Task 9

Create:

def is_even(number):

Return:

True

if even.

Return:

False

if odd.

Example:

print(is_even(4))

Output:

True
'''

def is_even(number):
    if number % 2 == 0:
        return 'Even'
    else:
        return 'Odd'

print(is_even(5))
```

### Level 4: Combine Functions + Conditions

```python
'''
Task 10

Create:

def grade(marks):

Return:

A
B
C
D
F

based on marks.

Example:

print(grade(85))

Output:

B
'''

def grade(marks):
    if 0 < marks > 100:
        return 'Enter valid marks'
    elif marks > 89:
        return 'A'
    elif marks > 79:
        return 'B'
    elif marks > 69:
        return 'C'
    elif marks > 59:
        return 'D'
    elif marks > 49:
        return 'E'
    elif marks > 39:
        return 'F'
    else:
        return 'Fail'

print(grade(85))
```

```python
'''
Task 11

Create:

def login(username, password):

Correct values:

anish
1234

Return:

Login Successful

or

Invalid Login
'''

def login(username, password):
    username = "anish"
    password = 1234

    if username == username and password == password:
        return 'Login Successful'
    else:
        return 'Invalid Login'

print(login('anish', 1234))
```

```python
'''
Challenge 1

Create:

def calculator(a, b, operation):

Examples:

calculator(10, 5, "+")

Output:

15

Examples:

calculator(10, 5, "-")

Output:

5

Support:

+
-
*
/

Use if, elif, else.
'''

def calculator(a, b, operation):
    if operation == "+":
        return a + b
    elif operation == "-":
        return a - b
    elif operation == "*":
        return a * b
    elif operation == "/":
        return a / b
    else:
        return 'Invalid Operation'

calculator(10, 5, "+")
```

```python
'''
Challenge 2 (Most Important)

Create:

def student_result(name, marks):

Return:

Name
Grade
Pass/Fail

Example:

student_result("Anish", 85)

Output:

Student: Anish
Grade: B
Status: Pass
'''

def student_result(name, marks):
    if 0 < marks > 100:
        return 'Enter valid marks'
    elif marks >= 90 and marks <= 100:
        grade = 'A'
    elif marks >= 80:
        grade = 'B'
    elif marks >= 70:
        grade = 'C'
    elif marks >= 60:
        grade = 'D'
    elif marks >= 50:
        grade = 'E'
    else:
        grade = 'F'

    if marks >= 50:
        status = "Pass"
    else:
        status = "Fail"
        
    print("Student:", name)
    print("Grade:", grade)
    print("Status:", status)

student_result("Anish", 85)
```

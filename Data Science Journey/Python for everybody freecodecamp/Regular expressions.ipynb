```python
'''
Exercise 1 – Find Numbers
text = "John is 25 years old and has 3 dogs."

Output:

['25', '3']
'''

import re

text = "John is 25 years old and has 3 dogs."
numbers = re.findall(r'[0-9]+', text)
print(numbers)
```

```python
import re

text = "John is 25 years old and has 3 dogs."

numbers = re.findall(r"\d+", text)

print(numbers)
```

```python
'''
Exercise 2 – Find Email Addresses
text = """
john@gmail.com
alice@yahoo.com
invalid-email
bob@hotmail.com
"""

Extract only the valid email addresses.
'''

import re

text = """
john@gmail.com
alice@yahoo.com
invalid-email
bob@hotmail.com
"""

valid_email = re.findall(r"\S+@\S+\.\S+", text)
print(valid_email)
```

```python
'''
Exercise 3 – Replace Digits

Replace every number with "X".

Example:

Order 12345

↓

Order X
'''

import re

order = "Order 12345"

new_line = re.sub(r'\d+', 'X', order)

print(new_line)
```

```python
'''
Exercise 4 – Extract Dates
Meeting: 15/06/2026
Birthday: 22/10/1998

Return the dates.
'''

dates = ''' 
Meeting: 15/06/2026
Birthday: 22/10/1998
'''

import re

date = re.findall(r"\d+/\d+/\d+", dates)

print(date)
```

```python
'''
Exercise 5 – Validate Phone Number

Ask the user for a phone number.

Accept only 10 digits.
'''

import re

def validate_number():

    # asking for user input
    number = input("Enter your phone number: ")

    # pattern for exactly 10 numbers
    pattern = r"^\d{10}$"

    # matching pattern and input
    if re.match(pattern, number):
        print("Valid number")
    else:
        print("Invalid number")

validate_number()
```

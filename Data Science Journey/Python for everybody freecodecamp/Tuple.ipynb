```python
'''
Task 7: Function Returning a Tuple

Create:

def calculate(a, b):

Return:

Sum
Difference
Product

Example:

result = calculate(10, 5)

Expected output:

15
5
50

Use tuple unpacking.
'''

def calculate(a, b):
    total = a + b
    difference = a - b
    product = a * b

    return total, difference, product

total, difference, product = calculate(10, 5)

print(total)
print(difference)
print(product)
```

```python
'''
Task 8: Largest Number

Without using max():

numbers = (12, 45, 7, 89, 23)

Find the largest value.
'''

numbers = (12, 45, 7, 89, 23)

def max_number(numbers):

    max_num = None

    for number in numbers:
        if max_num is None or number > max_num:
            max_num = number
    return max_num

print(max_number(numbers))
```

```python
'''
ask 9: Average

Without using sum():

numbers = (10, 20, 30, 40, 50)

Return the average.
'''

numbers = (10, 20, 30, 40, 50)

def avg_num(numbers):
    if len(numbers) == 0:
        return None
        
    total = 0

    for number in numbers:
        total += number

    average = total / len(numbers)

    return average

print("Average =", avg_num(numbers))
```

```python
'''
Task 10: Mini Challenge

Create a function:

def statistics(numbers):

Return:

Largest
Smallest
Average

Example:

largest, smallest, average = statistics((10, 20, 30, 40, 50))

Output:

Largest: 50
Smallest: 10
Average: 30
'''

# function to collect numbers and to create tuple
def get_numbers():

    # Initializing empty list
    numbers = []

    # asking for manadatory first input
    while True:

        value = input("Enter a number (or 'done'): ")

        if value.lower() == "done":
            return()

        try:
            numbers.append(int(value))
            break
            
        except ValueError:
            print("Please enter at least one valid number!")

    
    # asking for optional input until user inputs done
    while True:

        value = input("Enter a number (or 'done'): ")

        # exit condition
        if value == "" or value.lower() == "done":
            break

        #safe conversion
        try:
            #appending value entered to the numbers list
            numbers.append(int(value))
        except ValueError:
            print("Please enter a valid number!")

    # creating numbers tuple from numbers list
    return tuple(numbers)


# function to do calculations
def statistics(numbers):

    # checking whether numbers is empty or not. This is safe guarding 
    if len(numbers) == 0:
        return None
        
    total = 0
    largest = None
    smallest = None

    for number in numbers:
        
        if largest is None or number > largest:
            largest = number
            
        if smallest is None or number < smallest:
            smallest = number

        total += number
        
    average = total / len(numbers)
    
    return largest, smallest, average

# calling get_numbers function
numbers = get_numbers()

result = statistics(numbers)

# handling error if user simply press enter at the begining
if result is None:
    print("No numbers were entered.")
else:
    # unpacking tuples
    largest, smallest, average = result

    print("Largest:", largest)
    print("Smallest:", smallest)
    print("Average", average)
```

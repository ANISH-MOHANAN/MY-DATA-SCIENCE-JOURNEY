### elif

```python
'''Task 4: Grade Calculator

Input marks.

Rules:

90+ → A
80-89 → B
70-79 → C
60-69 → D
Below 60 → F
'''
marks = int(input("Enter your marks: "))

if marks >= 90:
    print("A")
elif marks >= 80 and marks <= 89:
    print("B")
elif marks >= 70 and marks <= 79:
    print("C")
elif marks >= 60 and marks <= 69:
    print("D")
else:
    print("F")
```

```python
'''
Task 6: BMI Category

Input BMI value.

Example categories:

BMI < 18.5 → Underweight
BMI < 25 → Normal
BMI < 30 → Overweight
Otherwise → Obese
'''

bmi_value = float(input("Enter your BMI value: "))

if bmi_value < 18.5:
    print("Underweight")
elif bmi_value >= 18.5 and bmi_value < 25:
    print("Normal")
elif bmi_value >= 25 and bmi_value < 30:
    print("Overweight")
else:
    print("Obese")
```

### Nested Conditions

```python
'''
Task 7: Login System

Create:

username = "anish"
password = "1234"

Ask user to enter both.

Check:

Username correct?
Password correct?

Print appropriate messages.
'''

username = "anish"
password = "1234"

login_name = input("Enter your username: ")
login_pass = input("Rnter your password: ")

if login_name == username:
    if login_pass == password:
        print("Login successful")
        print("Welcome Anish")
    else:
        print("Wrong password. Try again.")
else:
    print("Username not found")
```

### try / except

```python
try:
    number = int(input("Enter a number: "))
    print("You entered: ", number)
    
    square = number * number
    print("Square is: ", square)
    
except valueError:
    print("Please enter a valid number")
   
```

```python
'''
Task 10: Safe Division

Ask:

numerator
denominator

Handle:

10 / 0

using:

ZeroDivisionError

Example:

Cannot divide by zero
'''
numerator = int(input("Enter the numerator value: "))
denominator = int(input("Enter the denominator value: "))

try:
    division_result = numerator / denominator
    print("The result of division is: ", division_result)

except ZeroDivisionError:
    print("Cannot divide by zero")
    
```

```python
'''
Task 11: Multiple Error Handling

Handle both:

Invalid number
Divide by zero

using separate except blocks.
'''

try:
    numerator = int(input("Enter the numerator value: "))
    denominator = int(input("Enter the denominator value: "))
    
    division_result = numerator / denominator
    print("The result of division is: ", division_result)

except ValueError:
    print("Please enter a valid number")
    
except ZeroDivisionError:
    print("Cannot divide by zero")
```

```python
'''
Challenge 2

Simple ATM

Menu:

1. Check Balance
2. Deposit
3. Withdraw

Use:

if
elif
else
'''
balance = 1000

try:
    print("ATM Menu")
    print("1. Check Balance")
    print("2. Deposit")
    print("3. Withdraw")
    
    choice = int(input("Enter your choice from (1-3): "))
    
#checking balance    
    if choice == 1:
        print("Your balance is: ", balance)

#deposit
    elif choice == 2:
        amount = float(input("Enter deposit amount: "))
        balance = balance + amount
        print("Deposit successful")
        print("New balance: ", balance)
    
#deposit
    elif choice == 3:
        amount = float(input("Enter withdrawal amount: "))
    
        if amount <= balance:
            balance = balance - amount
            print("Withdrawal successful")
            print("Remaining balance", balance)
        else:
            print("Insufficient balance")
    
    else:
        print("Invalid choice")

except ValueErorr:
    print("Please enter numbers only")
```

```python
'''
Challenge 3

Student Result System

Input:

Name
Marks

Print:

Grade
Pass/Fail

Use:

if
elif
else
try/except
'''

try:
    name = input("Enter your name: ")
    marks = int(input("Enter your marks: "))

    if marks < 0 or marks > 100:
        print("Marks should be between 0 and 100")
    elif marks >= 90:
        grade = "A"
        status = "Pass"
    elif marks >= 80: 
        grade = "B"
        status = "Pass"
    elif marks >= 70:
        grade = "C"
        status = "Pass"
    elif marks >= 60:
        grade = "D"
        status = "Pass"
    elif marks >= 50:
        grade = "E"
        status = "Pass"
    else:
        grade = "F"
        status = "Fail"
        
    if 0 <= marks <= 100:
        print("Student: ", name)
        print("Marks: ", marks)
        print("Grade: ", grade)
        print("Status: ", status)

        
except ValueError:
    print("Enter valid numbers only")
```

```python
'''
Mini Challenge – Student Records

Create a program with this menu:

1. Add Student
2. Show Students
3. Exit

When adding a student:

Enter name:
Enter mark:

Store the data in:

students.txt

Example file:

John,82
Alice,95
Bob,74

When the user selects Show Students, display all records.
'''

# menu function
def menu():
    print("\n1. Add Student")
    print("2. Show Students")
    print("3. Exit")


#function to validtae marks
def get_valid_mark():

    while True:

        mark = input("Enter mark: ")

        # Check whether mark is numeric
        if not mark.isdigit():
            print("Marks must contain only numbers.")
            continue

        mark = int(mark)

        # Check range of mark, Mark must be between 0-100
        if 0 <= mark <= 100:
            return mark

        print("Marks must be between 0 and 100.")


# Function to check duplicates
def student_exists(name):

    try:

        # Opening file in read mode to check for duplicates
        with open("students.txt", "r", encoding="utf-8") as file:

            # looping through the file to get details
            for student in file:

                # Remove newline
                student = student.strip()

                # Skip empty lines
                if student == "":
                    continue

                existing_name, existing_mark = student.split(",")

                if existing_name.lower() == name.lower():
                    return True

    except FileNotFoundError:
        return False

    return False


# Function to add student
def add_student():

    name = input("\nEnter name: ").title()

    # Check duplicate
    if student_exists(name):
        print("Student already exists.")
        return

    mark = get_valid_mark()

    with open("students.txt", "a", encoding="utf-8") as file:
        file.write(f"{name},{mark}\n")

    print("Student added successfully.")



# Function to show students
def show_students():

    try:

        with open("students.txt", "r", encoding="utf-8") as file:

            # reading file contents to students variable
            students = file.readlines()

            # if students is empty
            if not students:
                print("\nNo student records found.")
                return

            print("\nStudent Records")
            print("-" * 30)
            print(f'{"Name":<20}{"Mark"}')
            print("-" * 30)

            for student in students:

                student = student.strip()

                if student == "":
                    continue

                name, mark = student.split(",")

                print(f"{name:<20}{mark}")

    except FileNotFoundError:
        print("\nNo student records found.")

    
    
# Main Program
def student_record():

    while True:

        menu()

        # asing to enter option from menu and handling error if alphabets are entered
        try:
            option = int(input("\nSelect an option: "))

        except ValueError:
            print("Please enter numbers only.")
            continue

        # Function call to add student details if option 1 is selected
        if option == 1:
            add_student()

        # Function call to show student details if option 2 is selected
        elif option == 2:
            show_students()

        # Exit if option 3 is selected
        elif option == 3:
            print("Thank you for using Student Records.")
            break

        # Asking for valid option if nothing from 1-3 is not entered 
        else:
            print("Please choose a valid option.")


student_record()
```
